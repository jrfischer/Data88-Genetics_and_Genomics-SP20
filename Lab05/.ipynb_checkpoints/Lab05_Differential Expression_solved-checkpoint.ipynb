Today we're going to explore one of the most central topics in not only genomics, but all of statistics--hypothesis testing. In particular, we're interested in testing whether two populations have different means. ## Part I : The t-test and rank-sum test for differences in means

The t-test is used to test whether there is evidence that the means of two populations are different. It assumes that both populations are normally distributed. Depending on the variation, it may also assume that the variances are equal. Today we're going to explore the most common version of the t-test with normal data and then show it how it fails when the data are not normal.

```python
# Simulate two groups of 50 observations from standard normal distributions (mean 0, sd 1)
# hint: np.random.normal(mean, sd, number)
norm_obs_1 = np.random.normal(0, 1, 50)
norm_obs_2 = np.random.normal(0, 1, 50)
```

Now you have two sets of 50 normal RVs and want to test whether the means of the populations they come from are different. To start, determine the null and alternative hypotheses below:

null hypothesis: mean_1 == mean_2
    
alternative hypothesis: mean_1 != mean_2

```python
# Let's run the t-test on our data and return the p-value. What do you expect on average?
# hint: stats.ttest_ind(group1, group2)[1] returns the p-value. [0] returns the test statistic
stats.ttest_ind(norm_obs_1, norm_obs_2)[1]
```

One iteration can only tell us so much. Let's run it a bunch of times to see what happens! We want to make sure our test isn't too sensitive when the null is true. Based on the histogram, how often do we expect to incorrectly reject the null?

```python
# Simulate 10000 sets of observations and do the t-test each time. Store the p-values so we can examine them.
# Initialize the list to store p-values.
# Use a for loop to iteratively generate the p-values. In each iteration, draw two groups of 
# 50 standard normal observations and perform a t-test. Store the p-value.
p_values_1 = []
for i in np.arange(0, 10000):
    norm_obs_1 = np.random.normal(0, 1, 50)
    norm_obs_2 = np.random.normal(0, 1, 50)
    p_values_1.append(stats.ttest_ind(norm_obs_1, norm_obs_2)[1])
    
# What does the distribution of p-values look like? Let's investigate by making a histogram.
# Use 20 bins and density = True
plt.hist(p_values_1, bins = 20, density = True)
plt.title('t-test, N(0,1) vs N(0,1)')
plt.xlabel('p-value')
plt.ylabel('Density')
plt.show()
```

How about when the means are different? The ability of a test to correctly reject the null is known as its power. Let's examine the behavior of the t-test for a moderate difference in means.

```python
# Same as the preceding cell, but with means of 0 and 0.2 for the two groups.
p_values_2 = []
for i in np.arange(0, 10000):
    norm_obs_1 = np.random.normal(0, 1, 50)
    norm_obs_2 = np.random.normal(.2, 1, 50)
    p_values_2.append(stats.ttest_ind(norm_obs_1, norm_obs_2)[1])
    
# What does the distribution of p-values look like?
plt.hist(p_values_2, bins = 20, density = True)
plt.title('t-test, N(0,1) vs N(0.2,1)')
plt.xlabel('p-value')
plt.ylabel('Density')
plt.show()
```

We've seen that the t-test performs sensibly when the data are normally distributed. Now let's examine what happens when we violate this assumption. The Cauchy distribution is a symmetric distribution that happens to have extremely heavy tails, meaning it frequently produces values far away from its center. Let's see how it looks--run the code below a few times and note what happens.

```python
# Let's compare histograms and density estimates for 1000 randomly sampled Cauchy and standard normal RVs.

plt.figure(figsize=(15,5))
plt.subplot(121)
sns.distplot(stats.cauchy.rvs(loc=0, scale=1, size=1000), kde=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Cauchy')

plt.subplot(122)
sns.distplot(np.random.normal(0, 1, 1000), kde=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Gaussian')

plt.show()
```

The t-test tests for differences in means, whereas the rank-sum test tests for something called "stochastic dominance". Loosely speaking, let's interpret this to mean a shift in the distribution. For symmetric distributions, this is equivalent to differences in means or medians. Let's evaluate how the p-values for two two tests differ on normal- and Cauchy-distributed data. (Note: the rank-sum test has a few names. These include the Mann-Whitney U test and the Wilcoxon rank-sum test. I usually just say "rank-sum test".)

```python
# t-test on Gaussian and Cauchy data. Try changing the location parameter and seeing what happens
location = .2
n_trials = 10000
t_p_normal_vec = []
t_p_cauchy_vec = []
for _ in np.arange(n_trials):
    normal_obs_1 = np.random.normal(0, 1, 50)
    normal_obs_2 = np.random.normal(location, 1, 50)
    cauchy_obs_1 = stats.cauchy.rvs(loc=0, scale=1, size=50)
    cauchy_obs_2 = stats.cauchy.rvs(loc=location, scale=1, size=50)
    t_p_normal_vec.append(stats.ttest_ind(normal_obs_1, normal_obs_2)[1])
    t_p_cauchy_vec.append(stats.ttest_ind(cauchy_obs_1, cauchy_obs_2)[1])

plt.figure(figsize=(15,5))
plt.hist(t_p_normal_vec, bins = 20, density = True, color = 'Blue', alpha = 0.5)
plt.hist(t_p_cauchy_vec, bins = 20, density = True, color = 'Red', alpha = 0.5)
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(['Gaussian','Cauchy'])
plt.title('t-test')
plt.show()
```

```python
# Repeat the analysis but replace the t-test with the rank-sum test
# hint stats.mannwhitneyu(x = group1, y = group2)[1] returns the p-value
# from the rank-sum test applied to group1 and group2

location = .2
rs_p_normal_vec = make_array()
rs_p_cauchy_vec = make_array()
for _ in np.arange(1000):
    normal_obs_1 = np.random.normal(0, 1, 50)
    normal_obs_2 = np.random.normal(location, 1, 50)
    cauchy_obs_1 = stats.cauchy.rvs(loc=0, scale=1, size=50)
    cauchy_obs_2 = stats.cauchy.rvs(loc=location, scale=1, size=50)
    rs_p_normal_vec = np.append(rs_p_normal_vec, stats.mannwhitneyu(x=normal_obs_1, y=normal_obs_2)[1])
    rs_p_cauchy_vec = np.append(rs_p_cauchy_vec, stats.mannwhitneyu(x=cauchy_obs_1, y=cauchy_obs_2)[1])

plt.figure(figsize=(15,5))
plt.hist(rs_p_normal_vec, bins = 20, density = True, color = 'Blue', alpha = 0.5)
plt.hist(rs_p_cauchy_vec, bins = 20, density = True, color = 'Red', alpha = 0.5)
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(['Gaussian','Cauchy'])
plt.title('Rank-sum test')
plt.show()
```

So it seems that the rank-sum test is robust to the choice of distribution while the t-test suffers quite a bit when our assumptions are invalid. 

## Part II: Permutation Tests 

Permutation tests provide another way to perform hypothesis testing. They are remarkably flexible, but can be computationally demanding. One benefit is the transparent way in which randomness enters the procedure. Let's see how they work when testing for a difference in medians.

### Difference in Medians


```python
# Write a function that takes two arrays as input and performs a permutation test for the difference of medians.

def medians_permutation_test(array_1, array_2, num_perm = 1000, test_type = 'two-sided'):
    true_diff = np.median(array_1) - np.median(array_2)
    null_diffs = make_array()
    merged_array = np.append(array_1, array_2)
    for i in np.arange(num_perm):
        shuffled_array = random.sample(list(merged_array), k = len(merged_array))
        group_1 = shuffled_array[:len(array_1)]
        group_2 = shuffled_array[len(array_1):]
        diff_medians = np.median(group_1) - np.median(group_2)
        null_diffs = np.append(null_diffs, diff_medians)
    if test_type == 'two-sided':
            p_val = np.sum(np.abs(null_diffs) >= np.abs(true_diff))/num_perm
    elif test_type == 'greater':
            p_val = np.sum(null_diffs >= true_diff)/num_perm
    elif test_type == 'less':
            p_val = np.sum(null_diffs <= true_diff)/num_perm
    else:
        return "test_type improperly specified; choices are 'two-sided', 'greater', 'less'"
        
    return [p_val, null_diffs, true_diff]
        
```

### Normal example

```python
# Let's start with observations from the normal distribution. Sample two groups of size 50 from 
# normal distributions with standard deviation 1 but respective means of 0 and 0.2.
normal_perm = medians_permutation_test(array_1 = np.random.normal(0,1,50), array_2 = np.random.normal(.2, 1, 50))
```

```python
# Let's make a histogram of the null distribution with our true value displayed
plt.hist(normal_perm[1], density = True)
plt.axvline(x=normal_perm[2], color = 'red', linewidth = 2)
plt.xlabel('Difference in medians')
plt.ylabel('Density')
plt.title('Perm. test, Gaussian. p-value is ' + str(normal_perm[0]))
plt.show()
```

### Cauchy example

```python
# Repeat with cauchy distributions with scale paramter 1 but location parameters of 0 and 0.2.
cauchy_perm = medians_permutation_test(array_1 = stats.cauchy.rvs(loc=0, scale=1, size=50),
                                       array_2 = stats.cauchy.rvs(loc=.2, scale=1, size=50))
```

```python
# Produce the histogram as before
plt.hist(cauchy_perm[1], density = True)
plt.axvline(x=cauchy_perm[2], color = 'red', linewidth = 2)
plt.xlabel('Difference in medians')
plt.ylabel('Density')
plt.title('Perm. test, Cauchy. p-value is ' + str(cauchy_perm[0]))
plt.show()
```

```python
gene_p_vals = []
for x in np.arange(2, female_data.shape[1]):
    female_gene_data = female_data.iloc[:,x]
    male_gene_data = male_data.iloc[:,x]
    p_val = stats.mannwhitneyu(female_gene_data, male_gene_data)[1]
    gene_p_vals.append(p_val)
    
len(gene_p_vals)
```

```python
# Compute the number of genes which are differentially expressed (p-value less than or equal to 0.05)
n_genes_de = np.sum(np.array(gene_p_vals) <= 0.05) # you may need to apply np.array to your p-values if you stored them in a list
n_genes_de
```

## Part IV:  Power comparison of paired and unpaired permutation test for bivariate Gaussians

```python
# Write a function that performs a paired and unpaired permutation test on bivariate Gaussians.
# Have the function take 1 input -- the mean of the second coordinate

def power_comparison(mu_y):
    p_values_unpaired = make_array()
    p_values_paired = make_array()
    for j in np.arange(100):
        obs = multivariate_normal.rvs([0, mu_y], [[1, .5], [.5,1]], size=15)
        x = obs[:,0]
        y = obs[:,1]
        true_means_diff = np.mean(x) - np.mean(y)
        null_diffs_unpaired = make_array()
        null_diffs_paired = make_array()
        merged_obs = np.append(x,y)
        for i in np.arange(100):
            shuffled_obs = random.sample(list(merged_obs), k = len(merged_obs))
            x_new = shuffled_obs[:len(x)]
            y_new = shuffled_obs[len(x):]
            null_diffs_unpaired = np.append(null_diffs_unpaired, np.mean(x_new) - np.mean(y_new))
            null_diffs_paired = np.append(null_diffs_paired, np.mean(random.choices([-1,1], k = 15) * (x-y)))
        p_values_unpaired = np.append(p_values_unpaired, np.mean(abs(null_diffs_unpaired) >= abs(true_means_diff)))
        p_values_paired = np.append(p_values_paired, np.mean(abs(null_diffs_paired) >= abs(true_means_diff)))
    return [np.mean(p_values_unpaired <= 0.05), np.mean(p_values_paired <= 0.05)]
```

```python
unpaired_power = make_array()
paired_power = make_array()
mus = np.array([-2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2])
for i in mus:
    power_res = power_comparison(i)
    unpaired_power = np.append(unpaired_power, power_res[0])
    paired_power = np.append(paired_power, power_res[1])
```

```python
plt.figure()
plt.plot(mus, unpaired_power)
plt.plot(mus, paired_power)
plt.legend(['Unpaired', 'Paired'])
plt.xlabel('Mean of Distribution')
plt.ylabel('Power')
plt.title('Comparison of permutation tests')
plt.show()
```
